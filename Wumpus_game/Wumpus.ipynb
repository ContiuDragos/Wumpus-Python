```python
import numpy as np
import random as random
```

```python
class Jucator:
    def __init__(self):
        self.gloante=1
        self.areComoara=0
        self.pozitie=[0,0]
class Wump:
    def __init__(self):
        self.pozitie=[0,0]
        self.viata=1
```

```python
def puneEroul(h):
    h[0][0]=1
def puneComoara(h):
    x=random.randint(0,9)
    y=random.randint(0,9)
    if(h[x][y]!=0):
        y=(y*2)%10
    h[x][y]=5
def puneCapcana(h):
    x=random.randint(0,9)
    y=random.randint(0,9)
    while(h[x][y]!=0):
        x=(x*2)%10
        y=(y*2)%10
    h[x][y]=6
    if(x>0 and h[x-1][y]==0):
        h[x-1][y]=7
    if(x<9 and h[x+1][y]==0):
        h[x+1][y]=7
    if(y<9 and h[x][y+1]==0):
        h[x][y+1]=7
    if(y>0 and h[x][y-1]==0):
        h[x][y-1]=7
    if((x==0 and y==1) or (x==1 and y==0)):
        print("VEZI SA NU CAZI IN GROAPA MAN!")
def puneWumpus(h,wumpus):
    x=random.randint(0,9)
    y=random.randint(0,9)
    while(h[x][y]!=0):
        y=(y*2)%10
        x=(x*2)%10
    h[x][y]=9
    wumpus.pozitie=[x,y]
    if(x>0 and h[x-1][y]==0):
        h[x-1][y]=8
    if(x<9 and h[x+1][y]==0):
        h[x+1][y]=8
    if(y>0 and h[x][y-1]==0):
        h[x][y-1]=8
    if(y<9 and h[x][y+1]==0):
        h[x][y+1]=8
    if((x==0 and y==1) or (x==1 and y==0)):
        print("VEZI CA E BAU BAU LANGA TINE!")
def afiseazaControale():
    print("CONTROALE:\nW-mergi in sus\nS-mergi in jos\nA-mergi in stanga\nD-mergi in dreaota\nQ-trage cu pistolul\n")   
def afiseazaLegenda():
    print("1-Eroul\n5-Comoara\n6-Capcana\n7-Briza de langa capcana\n8-Mirosul Wumpusului\n9-Wumpusul")
    print("Scopul este sa gasesti comoara si sa omori Wumpusul. Bafta!\n")
```

```python
def verificaMiscarea(h,wumpus,c):
    if(c=='w' and wumpus.pozitie[0]>0 and (h[wumpus.pozitie[0]-1][wumpus.pozitie[1]]==8 or h[wumpus.pozitie[0]-1][wumpus.pozitie[1]]==7)):
            return 1
    elif(c=='s' and wumpus.pozitie[0]<9 and (h[wumpus.pozitie[0]+1][wumpus.pozitie[1]]==8 or h[wumpus.pozitie[0]+1][wumpus.pozitie[1]]==7)):
            return 1
    elif(c=='a' and wumpus.pozitie[1]>0 and (h[wumpus.pozitie[0]][wumpus.pozitie[1]-1]==8 or h[wumpus.pozitie[0]][wumpus.pozitie[1]-1]==7)):
            return 1
    elif(c=='d' and wumpus.pozitie[1]<9 and (h[wumpus.pozitie[0]][wumpus.pozitie[1]+1]==8 or h[wumpus.pozitie[0]][wumpus.pozitie[1]+1]==7)):
            return 1
    else:
         return 0
def refaPozitieWumpus(h,wumpus,var):
    if(var=='w'):
        wumpus.pozitie[0]-=1
        h[wumpus.pozitie[0]][wumpus.pozitie[1]]=9
        h[wumpus.pozitie[0]+1][wumpus.pozitie[1]]=8
        if(wumpus.pozitie[0]>0 and h[wumpus.pozitie[0]-1][wumpus.pozitie[1]]==0):
            h[wumpus.pozitie[0]-1][wumpus.pozitie[1]]=8   
        if(wumpus.pozitie[1]>0 and h[wumpus.pozitie[0]][wumpus.pozitie[1]-1]==0):
            h[wumpus.pozitie[0]][wumpus.pozitie[1]-1]=8
        if(wumpus.pozitie[0]>0 and wumpus.pozitie[1]<9 and h[wumpus.pozitie[0]-1][wumpus.pozitie[1]+1]==0):
            h[wumpus.pozitie[0]][wumpus.pozitie[1]+1]=8
        if(wumpus.pozitie[0]<8 and h[wumpus.pozitie[0]+2][wumpus.pozitie[1]]==8):
            h[wumpus.pozitie[0]+2][wumpus.pozitie[1]]=0
        if(wumpus.pozitie[0]<9 and wumpus.pozitie[1]>0 and h[wumpus.pozitie[0]+1][wumpus.pozitie[1]-1]==8):
            h[wumpus.pozitie[0]+1][wumpus.pozitie[1]-1]=0
        if(wumpus.pozitie[0]<9 and wumpus.pozitie[1]<9 and h[wumpus.pozitie[0]+1][wumpus.pozitie[1]+1]==8):
            h[wumpus.pozitie[0]+1][wumpus.pozitie[1]+1]=0
    elif(var=='s'):
        wumpus.pozitie[0]+=1
        h[wumpus.pozitie[0]][wumpus.pozitie[1]]=9
        h[wumpus.pozitie[0]-1][wumpus.pozitie[1]]=8
        if(wumpus.pozitie[0]<9 and h[wumpus.pozitie[0]+1][wumpus.pozitie[1]]==0):
            h[wumpus.pozitie[0]+1][wumpus.pozitie[1]]=8
        if(wumpus.pozitie[1]>0 and h[wumpus.pozitie[0]][wumpus.pozitie[1]-1]==0):
            h[wumpus.pozitie[0]][wumpus.pozitie[1]-1]=8
        if(wumpus.pozitie[1]<9 and h[wumpus.pozitie[0]][wumpus.pozitie[1]+1]==0):
            h[wumpus.pozitie[0]][wumpus.pozitie[1]+1]=8
        if(wumpus.pozitie[0]>1 and h[wumpus.pozitie[0]-2][wumpus.pozitie[1]]==8):
            h[wumpus.pozitie[0]-2][wumpus.pozitie[1]]=0
        if(wumpus.pozitie[0]>0 and wumpus.pozitie[1]>0 and h[wumpus.pozitie[0]-1][wumpus.pozitie[1]-1]==8):
            h[wumpus.pozitie[0]-1][wumpus.pozitie[1]-1]=0
        if(wumpus.pozitie[0]>0 and wumpus.pozitie[1]<9 and h[wumpus.pozitie[0]-1][wumpus.pozitie[1]+1]==8):
            h[wumpus.pozitie[0]-1][wumpus.pozitie[1]+1]=0
    elif(var=='a'):
        wumpus.pozitie[1]-=1
        h[wumpus.pozitie[0]][wumpus.pozitie[1]]=9
        h[wumpus.pozitie[0]][wumpus.pozitie[1]+1]=8
        if(wumpus.pozitie[0]<9 and h[wumpus.pozitie[0]+1][wumpus.pozitie[1]]==0):
            h[wumpus.pozitie[0]+1][wumpus.pozitie[1]]=8
        if(wumpus.pozitie[0]>0 and h[wumpus.pozitie[0]-1][wumpus.pozitie[1]]==0):
            h[wumpus.pozitie[0]-1][wumpus.pozitie[1]]=8
        if(wumpus.pozitie[1]>0 and h[wumpus.pozitie[0]][wumpus.pozitie[1]-1]==0):    
            h[wumpus.pozitie[0]][wumpus.pozitie[1]-1]=8
        if(wumpus.pozitie[1]<8 and h[wumpus.pozitie[0]][wumpus.pozitie[1]+2]==8):
            h[wumpus.pozitie[0]][wumpus.pozitie[1]+2]=0
        if(wumpus.pozitie[0]>0 and wumpus.pozitie[1]<9 and h[wumpus.pozitie[0]-1][wumpus.pozitie[1]+1]==8):
            h[wumpus.pozitie[0]-1][wumpus.pozitie[1]+1]=0
        if(wumpus.pozitie[0]<9 and wumpus.pozitie[1]<9 and h[wumpus.pozitie[0]+1][wumpus.pozitie[1]+1]==8):
            h[wumpus.pozitie[0]+1][wumpus.pozitie[1]+1]=0
    elif(var=='d'):
        wumpus.pozitie[1]+=1
        h[wumpus.pozitie[0]][wumpus.pozitie[1]]=9
        h[wumpus.pozitie[0]][wumpus.pozitie[1]-1]=8
        if(wumpus.pozitie[0]<9 and h[wumpus.pozitie[0]+1][wumpus.pozitie[1]]==0):
            h[wumpus.pozitie[0]+1][wumpus.pozitie[1]]=8
        if(wumpus.pozitie[1]>0 and h[wumpus.pozitie[0]][wumpus.pozitie[1]-1]==0):
            h[wumpus.pozitie[0]][wumpus.pozitie[1]-1]=8
        if(wumpus.pozitie[1]<9 and h[wumpus.pozitie[0]][wumpus.pozitie[1]+1]==0):    
            h[wumpus.pozitie[0]][wumpus.pozitie[1]+1]=8
        if(wumpus.pozitie[1]>0 and h[wumpus.pozitie[0]][wumpus.pozitie[1]-2]==8): 
            h[wumpus.pozitie[0]][wumpus.pozitie[1]-2]=0
        if(wumpus.pozitie[0]<9 and wumpus.pozitie[1]>0 and h[wumpus.pozitie[0]+1][wumpus.pozitie[1]-1]==8): 
            h[wumpus.pozitie[0]+1][wumpus.pozitie[1]-1]=0
        if(wumpus.pozitie[0]<0 and wumpus.pozitie[1]>0 and h[wumpus.pozitie[0]-1][wumpus.pozitie[1]-1]==8):
            h[wumpus.pozitie[0]-1][wumpus.pozitie[1]-1]=0
```

```python
def mutaWumpus(h,erou,wumpus):
    if(erou.pozitie[0]<wumpus.pozitie[0]):
        if(verificaMiscarea(h,wumpus,'w')==1):
            refaPozitieWumpus(h,wumpus,'w')
        else:
            if(erou.pozitie[1]<wumpus.pozitie[1]):
                if(verificaMiscarea(h,wumpus,'a')==1):
                    refaPozitieWumpus(h,wumpus,'a')
            else:
                if(verificaMiscarea(h,wumpus,'d')==1):
                    refaPozitieWumpus(h,wumpus,'d')
    elif(erou.pozitie[0]>wumpus.pozitie[0]):
        if(verificaMiscarea(h,wumpus,'s')==1):
            refaPozitieWumpus(h,wumpus,'s')
        else:
            if(erou.pozitie[1]<wumpus.pozitie[1]):
                if(verificaMiscarea(h,wumpus,'a')==1):
                    refaPozitieWumpus(h,wumpus,'a')
            else:
                if(verificaMiscarea(h,wumpus,'d')==1):
                    refaPozitieWumpus(h,wumpus,'d')
    else:
        if(erou.pozitie[1]<wumpus.pozitie[1]):
            if(verificaMiscarea(h,wumpus,'a')==1):
                    refaPozitieWumpus(h,wumpus,'a')
            else:
                if(verificaMiscarea(h,wumpus,'d')==1):
                    refaPozitieWumpus(h,wumpus,'d')
       
```

```python
def trageEroul(h,wumpus,erou):
    if(erou.gloante==0):
        print("AI MERS LA RAZBOI FARA GLOANTE! FUGI ACUM")
        return
    else:
        mov=input("IN CE DIRECTIE TRAGI?")
        erou.gloante=0
        if(mov=='w'):
            if(erou.pozitie[1]==wumpus.pozitie[1] and erou.pozitie[0]>wumpus.pozitie[0]):
                omoaraWumpus(h,wumpus)
        elif(mov=='s'):
            if(erou.pozitie[1]==wumpus.pozitie[1] and erou.pozitie[0]<wumpus.pozitie[0]):
                omoaraWumpus(h,wumpus)
        elif(mov=='a'):
            if(erou.pozitie[0]==wumpus.pozitie[0] and erou.pozitie[1]>wumpus.pozitie[1]):
                omoaraWumpus(h,wumpus)
        elif(mov=='d'):
            if(erou.pozitie[0]==wumpus.pozitie[0] and erou.pozitie[1]<wumpus.pozitie[1]):
                omoaraWumpus(h,wumpus)
        if(wumpus.viata==1):
            print("MIE MI-AR FI RUSINE SA TRAG ASA!")

def omoaraWumpus(h,wumpus):
    wumpus.viata=0;
    print("L-AI NIMERIT! MIRA-M-AS")
    h[wumpus.pozitie[0]][wumpus.pozitie[1]]=0
    if(wumpus.pozitie[0]>0 and h[wumpus.pozitie[0]-1][wumpus.pozitie[1]]==8):
        h[wumpus.pozitie[0]-1][wumpus.pozitie[1]]=0
    if(wumpus.pozitie[0]<9 and h[wumpus.pozitie[0]+1][wumpus.pozitie[1]]==8):
        h[wumpus.pozitie[0]+1][wumpus.pozitie[1]]=0
    if(wumpus.pozitie[1]>0 and h[wumpus.pozitie[0]][wumpus.pozitie[1]-1]==8):
        h[wumpus.pozitie[0]][wumpus.pozitie[1]-1]=0
    if(wumpus.pozitie[1]<9 and h[wumpus.pozitie[0]][wumpus.pozitie[1]+1]==8):
        h[wumpus.pozitie[0]][wumpus.pozitie[1]+1]=0
    
```

```python
def adaugaMiscare(h,x,y):
    mov=input("TRAGI SAU MERGI? ")
    if(mov=='q' or mov=='Q'):
        return [-1,-1]
    if(mov=='w' or mov=='W'):
        if(x==0):
            print("CREZI CA POTI FACE ASTA? ITI ZIC EU CA NU POTI")
            return adaugaMiscare(x,y)
        else:
            return x-1,y
    if(mov=='s' or mov=='S'):
        if(x==9):
            print("CREZI CA POTI FACE ASTA? ITI ZIC EU CA NU POTI")
            return adaugaMiscare(x,y)
        else:
            return x+1,y
    if(mov=='a' or mov=='A'):
        if(y==0):
            print("CREZI CA POTI FACE ASTA? ITI ZIC EU CA NU POTI")
            return adaugaMiscare(x,y)
        else:
            return x,y-1
    if(mov=='d' or mov=='D'):
        if(y==9):
            print("CREZI CA POTI FACE ASTA? ITI ZIC EU CA NU POTI")
            return adaugaMiscare(x,y)
        else:
            return x,y+1
def refaGroapa(h,erou):
    h[erou.pozitie[0]][erou.pozitie[1]]=7
```

```python
def cautaComoara(h,erou,wumpus):
    langaGroapa=0
    if(h[erou.pozitie[0]][erou.pozitie[1]]==9):
        print("TE-A PAPAT WUMPUSUL TATAIE!")
        return
    if(erou.pozitie[0]==0 and erou.pozitie[1]==0 and erou.areComoara==1):
        if(wumpus.viata==0):
            print("AI CASTIGAT BAROSANE!")
            return
        else:
            print("TREBUIE SA OMORI SI WUMPUSUL!")
    if(h[erou.pozitie[0]][erou.pozitie[1]]==5):
        print("AI GASIT COMOARA BAROSANE!\nACUM TREBUIE SA IESI PE UNDE AI INTRAT!")
        erou.areComoara=1
    if(h[erou.pozitie[0]][erou.pozitie[1]]==8):
        print("VEZI CA E BAU BAU LANGA TINE!")
    if(h[erou.pozitie[0]][erou.pozitie[1]]==6):
        print("CE TIA-AM ZIS EU? AI CAZUT IN GROAPA TINERE!")
        return
    if(h[erou.pozitie[0]][erou.pozitie[1]]==7):
        print("VEZI SA NU CAZI IN GROAPA MAN!")
        langaGroapa=1
    h[erou.pozitie[0]][erou.pozitie[1]]=1
    for i in range(10):
        print(*h[i],sep=" ")
    h[erou.pozitie[0]][erou.pozitie[1]]=0
    if(langaGroapa==1):
        refaGroapa(h,erou)
        langaGroapa=0
    a,b=adaugaMiscare(h,erou.pozitie[0],erou.pozitie[1])
    if(a==-1 and b==-1):
        trageEroul(h,wumpus,erou)
    else:
        erou.pozitie=[a,b]
    if(wumpus.viata==1):
        mutaWumpus(h,erou,wumpus)
    cautaComoara(h,erou,wumpus)
            
```

```python
harta=np.zeros((10,10),dtype=int)
erou=Jucator()
wumpus=Wump()
puneEroul(harta)
puneComoara(harta)
puneCapcana(harta)
puneWumpus(harta,wumpus)
afiseazaLegenda()
afiseazaControale()
cautaComoara(harta,erou,wumpus)
```
